```python
import pandas as pd
import numpy as np
import itertools
```

```python

x=[]
y=[]
p=[]
def get_max_length(N):
    pit = None
    list_set = None
    unique_list=None
    if(N>3):
        
        for i in range(2,N):
            x.append(i)

        #print(x)
       # h = list(itertools.combinations(x,2))
        #for k in h:
            #print(k)
        for i in zip(list(itertools.combinations(x,2))):
            y.append((np.abs(i[0][0]-1),np.absolute(i[0][0]-i[0][1]),np.abs(i[0][1]-N)))
        #print(y)
        for j in (y):
            #print(j)
            #print('*****')
            p.append(max(j))
            #print(p)
        list_set = set(p)
        return list(list_set)
    else:
        return [0]
    
    
      
        
       
        
```

```python
 pit_in=[]
def get_list_dist(pit):
   
    for j in range(len(pit)):
        #print(len(pit))
        #print(pit[j])
        jj=get_max_length(pit[j])
        for m in jj:
            pit_in.append(m)
    u_list = set(pit_in)
    return list(u_list)
    
   
```

```python
x = get_max_length(24)
for j in range(4):
    x=get_list_dist(x)
    j=j+1
    print(j)
print(x)
```
