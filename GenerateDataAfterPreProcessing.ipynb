```python

test=np.hstack([b,b,b,b])
```

```python
test=np.delete(test,[4,8,10],axis=1)
```

```python
test[1:,4]=np.diff(test[:,1],axis=0)
test[1:,5]=np.diff(test[:,2],axis=0)
test[1:,6]=np.diff(test[:,3],axis=0)
```

```python
def median_filter(x, n): 
    x=x.reshape(-1,1)
    y=np.hstack([x,x])
    for i in range(n//2, len(x)-n-n//2):
        y[i+n//2,1] = np.median(x[i:i+n])
    return y[:,1]
  
```

```python
test[:,7]=median_filter(test[:,4],3)
test[:,8]=median_filter(test[:,5],3)
test[:,9]=median_filter(test[:,6],3)
```

```python
test[:,10]=np.cumsum(test[:,7],axis=0)
test[:,11]=np.cumsum(test[:,8],axis=0)
test[:,12]=np.cumsum(test[:,9],axis=0)
```
